```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN

df = pd.read_csv("C:\\Users\\Pradheesha\\Downloads\\Netflix recommendation system\\processed_df.csv")

# User-level features
user_features = df.groupby('userId').agg({
    'rating': 'mean',
    'movieId': 'count'
}).rename(columns={'rating':'avg_rating','movieId':'num_movies'})
```

```python
# KMeans
kmeans = KMeans(
    n_clusters=5,
    n_init=10,
    random_state=42
)

user_features['cluster'] = kmeans.fit_predict(user_features)
```

```python
import pickle
with open('models/kmeans_model.pkl', 'wb') as f:
    pickle.dump(kmeans, f)

print("KMeans model saved")
```

```python
# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
user_features['db_cluster'] = dbscan.fit_predict(user_features)
```

```python
import pickle

with open("models/dbscan_model.pkl", "wb") as f:
    pickle.dump(dbscan, f)
print("DBScan model saved")
```
